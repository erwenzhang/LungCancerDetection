```python
import os, os.path
from shutil import copyfile

destination_folder_name = 'file_collection'
destination_folder = os.curdir + os.sep + destination_folder_name + os.sep
namespace_separator = '_'
file_suffix = '.dcm'

if not os.path.exists(destination_folder):
    os.mkdir(destination_folder)

foldername_list = [name for name in os.listdir(os.curdir) if os.path.isdir(name) and not name == destination_folder_name]
for i in range(len(foldername_list)):
    dir_list = [name for name in os.listdir(os.curdir + os.sep + foldername_list[i]) ]
    file_count = len(dir_list)
    file_in_the_middle = os.curdir + os.sep + foldername_list[i] + os.sep + dir_list[int(file_count / 2)]
    print("Moving File: " + file_in_the_middle)
    copyfile(file_in_the_middle, destination_folder + foldername_list[i] + file_suffix)
```
